```python
import pandas as pd
import numpy as np
```

```python
# Import raw data scraped from webiste
raw_data = pd.read_csv('injury_data.csv')
```

```python
raw_data.columns=raw_data.iloc[0]
```

```python
# Create a copy dataframe to perform data cleanup on
copy_df = raw_data
```

```python
# Fix dataframe headers
indexNames = copy_df[copy_df['Team'] == 'Team' ].index
copy_df.drop(indexNames , inplace=True)
copy_df.drop(columns=0 , inplace=True)
copy_df
```

```python
# Performing some regex to eliminate the bullet point in front of player names
copy_df["Relinquished"] = copy_df["Relinquished"].str.extract('\s(.*)')
copy_df["Acquired"] = copy_df["Acquired"].str.extract('\s(.*)')
copy_df
```

```python
# Check NaN count for each column
copy_df.isna().sum()
```

```python
# Convert NaNs in Acquired and Relinquished columns to empty strings and concatenate columns into new, Name column
for index, row in copy_df.iterrows():
    copy_df['Name'] = copy_df['Acquired'].fillna('') + copy_df['Relinquished'].fillna('')

copy_df
```

```python
# Drop now unnecessary Acquired and Relinquished columns
copy_df = copy_df.drop(['Acquired', 'Relinquished'], axis=1)
copy_df
```

```python
# Rearrange columns in DataFrame
copy_df = copy_df[['Name', 'Date', 'Notes', 'Team']]
copy_df
```

```python
# Save partially cleaned data to new csv
copy_df.to_csv("cleaned_injury_data.csv")
```

```python
### LOAD IN LIGHTLY CLEANED DATA TO DO MORE THOROUGH CLEANING ###
```

```python
raw_data = pd.read_csv('cleaned_injury_data.csv')
```

```python
# Turn raw_data into a dataframe
copy_2_df = raw_data
copy_2_df
```

```python
# Drop the Unnamed: 0 as it is an unnecessary false index
copy_2_df = copy_2_df.drop("Unnamed: 0",1)
copy_2_df
```

```python
# Separate the injury description from notes column
copy_2_df["Injury"] = copy_2_df["Notes"].str.extract('((?<=with ).*$)')
copy_2_df
```

```python
# Create new dataframe to clean pull info from 'Notes' column
injury_info_df = pd.DataFrame()

# extract length of DL/IL stint and place in a new column

# MLB changed name of DL to IL, so multiple statements are required handling each
injury_info_df["DL_Stint_Placed"] = copy_2_df["Notes"].str.extract('((?<=on ).*(?= DL))')
injury_info_df["IL_Stint_Placed"] = copy_2_df["Notes"].str.extract('((?<=on ).*(?= IL))')

# mulitple cases are also needed to handle notes that say either 'placed on' or 'activated from' DL/IL
injury_info_df["DL_Stint_Activated"] = copy_2_df["Notes"].str.extract('((?<=from ).*(?= DL))')
injury_info_df["IL_Stint_Activated"] = copy_2_df["Notes"].str.extract('((?<=from ).*(?= IL))')

injury_info_df
```

```python
# Replace NaNs with empty strings to prepare to concatenate values
injury_info_df = injury_info_df.fillna('')

# concatenate the DL/IL_stint_placed values into one placed_on column
injury_info_df['placed_on'] = injury_info_df.DL_Stint_Placed.str.cat(injury_info_df.IL_Stint_Placed)

# concatenate the DL/IL_stint_activated values into one activated_from column
injury_info_df['activated_from'] = injury_info_df.DL_Stint_Activated.str.cat(injury_info_df.IL_Stint_Activated)
injury_info_df
```

```python
# Copying injury_info_df dataframe into a new dataframe to offer as a checkpoint in case I mess up the dataframe
copy_df_3 = injury_info_df
```

```python
# drop the columns we used to concatenate earlier that are no longer necessary
copy_df_3 = copy_df_3.drop(["DL_Stint_Placed", "IL_Stint_Placed", "DL_Stint_Activated", "IL_Stint_Activated"],1)
copy_df_3
```

```python
# Create a new column to use for another round of concatenating
copy_df_3["Status"] = ''
copy_df_3
```

```python
# Replace empty strings with NaNs, then replace NaNs with '2' in order to make them easy to 
# find with conditional statement
copy_df_3 = copy_df_3.replace(r'^\s*$', np.nan, regex=True)
copy_df_3 = copy_df_3.fillna('2')
copy_df_3
```

```python
# loop through rows, updating status column with either "Placed" or "Activated" depending on transaction status
for index, row in copy_df_3.iterrows():
    if row["placed_on"] == '2':
        row["Status"] = 'Activated'
    if row["activated_from"] == '2':
        row["Status"] = 'Placed'

copy_df_3
```

```python
# go back and return all '2' values back to empty strings to prepare for one last concatenate
for index, row in copy_df_3.iterrows():
    if row["placed_on"] == '2':
        row["placed_on"] = ''
    if row["activated_from"] == '2':
        row["activated_from"] = ''

copy_df_3
```

```python
# concatenate 'placed_on' and 'activated_from' columns into new 'injury_length' column
copy_df_3['injury_length'] = copy_df_3.activated_from.str.cat(copy_df_3.placed_on)
copy_df_3
```

```python
# drop the now redundent 'placed_on' and 'activated_from' columns
copy_df_3 = copy_df_3.drop(['activated_from', 'placed_on'], 1)
copy_df_3
```

```python
# Join our copied dataframe with our originally pulled one 
copy_2_df = copy_2_df.join(copy_df_3)
```

```python
# Create a new dataframe with rearranged columns
cleaned_injury_df = copy_2_df[['Name', 'Team', 'Date', 'Status', 'injury_length', 'Injury']]
cleaned_injury_df
```

```python
cleaned_injury_df["location"] = ''
cleaned_injury_df
```

```python
cleaned_injury_df.dtypes
```

```python
# MLB changed name of DL to IL, so multiple statements are required handling each
cleaned_injury_df["right_location"] = cleaned_injury_df["Injury"].str.extract('(right)')
cleaned_injury_df["left_location"] = cleaned_injury_df["Injury"].str.extract('(left)')

cleaned_injury_df
```

```python
# Replace NaNs with empty strings to prepare to concatenate values
cleaned_injury_df = cleaned_injury_df.fillna('')

# concatenate the DL/IL_stint_placed values into one placed_on column
cleaned_injury_df['location'] = cleaned_injury_df.right_location.str.cat(cleaned_injury_df.left_location)

cleaned_injury_df
```

```python
# Drop now unnecessary columns after concatenating
cleaned_injury_df = cleaned_injury_df.drop(['right_location', 'left_location'], 1)
cleaned_injury_df
```

```python
# Create a list from the Injury column values to prepare to clean
list_of_Injury = cleaned_injury_df['Injury'].tolist()
list_of_Injury
```

```python
# Loop through list to delete the word 'right'
for i in range(len(list_of_Injury)):
    list_of_Injury[i] = list_of_Injury[i].replace('right', '')

list_of_Injury
```

```python
# Same as before but with 'left'
for i in range(len(list_of_Injury)):
    list_of_Injury[i] = list_of_Injury[i].replace('left', '')

list_of_Injury
```

```python
# Update Injury column with newly cleaned list
cleaned_injury_df['Injury'] = list_of_Injury
cleaned_injury_df
```

```python
cleaned_injury_df.to_csv("new_cleaned_injury_data.csv")
```
