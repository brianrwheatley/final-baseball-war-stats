```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Creating variables to help move from page to page via a new URL
# Website only loaded 25 items at a time, which is why j starts at 25
# k will increment each time a new page is loaded so our count includes a new set of table headers
i = 0
j = 25
k = 0

# Create a list to store the information scraped from the website
data = []
```

```python
# create a loop to update the URL after scraping the page
# the loop ends when we hit the last page, which would not have 25 items on it, causing j - i to not equal 25

while (j - i) == 25:
    URL = f"https://www.prosportstransactions.com/baseball/Search/SearchResults.php?Player=&Team=&BeginDate=2010-03-30&EndDate=2021-09-30&DLChkBx=yes&submit=Search&start={i}"    
    r = requests.get(URL) # send the GET request to the URL
    
    # locate the table from the html, store it in table and table_body variables
    soup = BeautifulSoup(r.content, 'html5lib')
    table = soup.find('table')
    table_body = table.find('tbody')
    
    # Grab data from table and put it into the list created above
    for tr in table_body.select('tr:has(td)'):
        tds = [td.get_text(strip=True) for td in tr.select('td')]
        data.append(tds)

    # Create dataframe from the data list created and appended above
    injury_data_df = pd.DataFrame(data)
    
    # increment i up by 25 so the URL updates and moves to the next page
    i = i + 25
    
    # update j to reflect the amount of rows currently in the dataframe
    # it is important this takes place before k is incremented by one, because header rows are not counted for URL
    j = (injury_data_df.index[-1] - k) + 25
    
    # increment k up by one, reflects how many pages we have scraped and accounts for a new header row for each scrape
    k = k + 1
    print(f"i = {i}")
    print(f"j = {j}")
    print(f"k = {k}")
    
```
