```python
import pandas as pd
warCSV = "Resources/jeffbagwell_war_historical.csv"
team_id_CSV = "Resources/team_codes.csv"

war_df = pd.read_csv(warCSV, encoding='ISO-8859-1')
teamID_df = pd.read_csv(team_id_CSV)
war_df
```

```python
war_df['team_ID'].unique()
```

```python
war_year_df = war_df[(war_df['year_ID'] >= 2010)]
war_year_df
```

```python
war_year_df['team_ID'].nunique()
# war_year_df
```

```python
teamID_present_df = teamID_df[(teamID_df['Last_Year'] >= 2010)]
teamID_present_df.count()['Team_ID']
```

```python
teamID_present_df = teamID_present_df.drop(['First_Year', 'Last_Year','Full_Team_Name'],axis=1)
# teamID_present_df
```

```python
cleaned_df = pd.merge(left=war_year_df,right=teamID_present_df,left_on='team_ID',right_on='Team_ID',how='left')
```

```python
filtered_df = cleaned_df.filter(['player_name','age','year_ID','team_ID','Team_Name','bwar162','pwar162','g_bat','pa','g_pitch','starts','innings','gms_P','gms_C','gms_1B','gms_2B','gms_3B','gms_SS','gms_LF','gms_CF','gms_RF','gms_OF','is_P'])
filtered_df
```

```python
pitchers_df = filtered_df[(filtered_df['is_P'] == 'Y')]
batters_df = filtered_df[(filtered_df['is_P'] == 'N')]
```

```python
pitchers_df = pitchers_df.filter(['player_name','age','year_ID','team_ID','Team_Name','pwar162','g_bat','pa','g_pitch','starts','innings','gms_P','is_P'])
pitchers_df.describe()
pitchers_df
```

```python
batters_df = batters_df.filter(['player_name','age','year_ID','team_ID','Team_Name','bwar162','g_bat','pa','gms_C','gms_1B','gms_2B','gms_3B','gms_SS','gms_LF','gms_CF','gms_RF','gms_OF','is_P'])
batters_df.describe()
batters_df
```
