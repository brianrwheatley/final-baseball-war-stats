```python
import pandas as pd
warCSV = "Resources/jeffbagwell_war_historical.csv"
team_id_CSV = "Resources/team_codes.csv"
injury_CSV = "Resources/final_injury_data.csv"

war_df = pd.read_csv(warCSV, encoding='ISO-8859-1')
teamID_df = pd.read_csv(team_id_CSV)
injury_df = pd.read_csv(injury_CSV)
war_df
```

```python
war_df['team_ID'].unique()
```

```python
war_year_df = war_df[(war_df['year_ID'] >= 2010)]
war_year_df
```

```python
war_year_df['team_ID'].nunique()
# war_year_df
```

```python
teamID_present_df = teamID_df[(teamID_df['Last_Year'] >= 2010)]
teamID_present_df.count()['Team_ID']
```

```python
teamID_present_df = teamID_present_df.drop(['First_Year', 'Last_Year','Full_Team_Name'],axis=1)
# teamID_present_df
```

```python
cleaned_df = pd.merge(left=war_year_df,right=teamID_present_df,left_on='team_ID',right_on='Team_ID',how='left')
```

```python
filtered_df = cleaned_df.filter(['player_name','age','year_ID','team_ID','Team_Name','bwar162','pwar162','g_bat','pa','g_pitch','starts','innings','gms_P','gms_C','gms_1B','gms_2B','gms_3B','gms_SS','gms_LF','gms_CF','gms_RF','gms_OF','is_P']).copy()
filtered_df['shortname'] = ""
filtered_df
```

```python
pitchers_df = filtered_df[(filtered_df['is_P'] == 'Y')]
batters_df = filtered_df[(filtered_df['is_P'] == 'N')]
```

```python
pitchers_df = pitchers_df.filter(['player_name','shortname','age','year_ID','team_ID','Team_Name','pwar162','g_bat','pa','g_pitch','starts','innings','gms_P','is_P'])
pitchers_df.describe()
p_short_df = pitchers_df['player_name'].str.split(' ', n=2, expand=True)
pitchers_df
# p_short_df
# pitchers_df.to_csv("pitchers_df.csv")
```

```python
p_shortname = (pd.Series(p_short_df[0].str[:1])).str.lower()
p_lastname = (pd.Series(p_short_df[1].str[:])).str.lower()
shortname = p_shortname + p_lastname
pitchers_df['shortname'] = shortname
pitchers_df
pitchers_df.to_csv("pitcher_war_data.csv")
```

```python
batters_df = batters_df.filter(['player_name','shortname','age','year_ID','team_ID','Team_Name','bwar162','g_bat','pa','gms_C','gms_1B','gms_2B','gms_3B','gms_SS','gms_LF','gms_CF','gms_RF','gms_OF','is_P'])
b_short_df = batters_df['player_name'].str.split(' ', n=2, expand=True)
# batters_df.describe()
batters_df
```

```python
b_shortname = (pd.Series(b_short_df[0].str[:1])).str.lower()
b_lastname = (pd.Series(b_short_df[1].str[:])).str.lower()
bat_shortname = b_shortname + b_lastname
batters_df['shortname'] = bat_shortname
batters_df
batters_df.to_csv("batter_war_data.csv")
```

```python
fullmerged_pitcher_df = pd.merge(injury_df, pitchers_df,  how='left', left_on=['shortname','year', 'Team'], right_on = ['shortname','year_ID', 'Team_Name'])
```
