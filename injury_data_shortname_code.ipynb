```python
import pandas as pd
import numpy as np
```

```python
raw_data = pd.read_csv('cleaned_injury_data_with_dates.csv')
```

```python
raw_data = raw_data.drop('Unnamed: 0', 1)
raw_data
```

```python
## create new dataframe from injury dataframe by splitting names into three columns
split_placed_df = raw_data['Name'].str.split(' ', n=2, expand=True)

## assigning strings from columns to new variable (first name, 1 character / full last name)
inj_shortname = (pd.Series(split_placed_df[0].str[:1])).str.lower()
inj_lastname = (pd.Series(split_placed_df[1].str[:])).str.lower()
shortname = inj_shortname + inj_lastname

## place new shortname code into injury dataframe
raw_data['shortname'] = shortname
raw_data
```

```python
raw_data.to_csv("final_injury_data.csv")
```
